Nessa etapa temos o pipeline único contendo pré-processamento e modelo , versionamento com timestamp para rollback , validação pós-salvamento e a separação  entre treino, avaliação e produção. A pipeline final esta versionada e o modelo pronto para uso em produção.

---



```python
import numpy as np
import joblib
import os
from datetime import datetime
```

```python
MODEL_DIR = "models"
MODEL_NAME = "fraud_model_pipeline"

os.makedirs(MODEL_DIR, exist_ok=True)
```

Essa  pipeline foi treinado e salvo inicialmente no Notebook 07

```python
pipeline_final = joblib.load("fraud_model_pipeline.joblib")
print("Pipeline carregado com sucesso")
```

```python
timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
model_version_path = f"{MODEL_DIR}/{MODEL_NAME}_v{timestamp}.joblib"

joblib.dump(pipeline_final, model_version_path)

print(f"Modelo versionado salvo em: {model_version_path}")
```

```python
model_loaded = joblib.load(model_version_path)
print("Modelo versionado carregado corretamente")
```


### Simulação de predição individual

---



Essa etapa simula uma única transação entrando no sistema antifraude

```python
X_sample = np.load("X_test.npy")[0].reshape(1, -1)

prob_fraude = model_loaded.predict_proba(X_sample)[0, 1]

threshold_operacional = 0.5  # exemplo, pode ser ajustado conforme Notebook 08
is_fraud = int(prob_fraude >= threshold_operacional)

print(f"Probabilidade de fraude: {prob_fraude:.4f}")
print(f"Decisão (0 = legítima, 1 = fraude): {is_fraud}")
```
